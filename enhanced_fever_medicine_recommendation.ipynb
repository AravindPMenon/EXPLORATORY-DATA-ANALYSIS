# ENHANCED FEVER MEDICINE RECOMMENDATION

### This dataset contains information on individuals with varying fever severities, including their age, gender, body mass index (BMI), and symptoms such as headache, body ache, and fatigue. It also includes lifestyle factors like physical activity, diet type, smoking history, and alcohol consumption, along with environmental data such as humidity and air quality index (AQI). The dataset records the heart rate, blood pressure, and chronic conditions of individuals, as well as the medications previously taken and those recommended based on their fever severity. This data can be analyzed to find patterns in medication recommendations and to study relationships between symptoms, lifestyle factors, and fever severity.

### 9)What are the most commonly recommended medications for each fever severity?

### 10)Are there any differences in recommended medications based on gender or age group?  

```python
bins = [0, 18, 60, 100]  # Define the age groups
labels = ['Child', 'Adult', 'Senior']  # Label the age groups
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Grouping by Age_Group and Recommended_Medication to count the occurrences
medication_by_age_group = df.groupby(['Age_Group', 'Recommended_Medication']).size().unstack().fillna(0)

# Print the result to check the medication recommendations based on age group
print(medication_by_age_group)
```

### 11)How often do previous medications match the recommended medications?

```python
matching_medications = df['Previous_Medication'] == df['Recommended_Medication']
match_percentage = (matching_medications.sum() / len(df)) * 100
print(f"Percentage of matching previous and recommended medications: {match_percentage:.2f}%")
```

### 12)Are symptoms like headache, body ache, and fatigue more prevalent in certain fever severity levels? 

```python
symptoms = ['Headache', 'Body_Ache', 'Fatigue']
symptom_prevalence = df.groupby('Fever_Severity')[symptoms].apply(lambda x: (x == 'Yes').mean() * 100)
print(symptom_prevalence)
```

### 13)How do chronic conditions and allergies affect the recommended medications? 

```python
medication_effects = df.groupby(['Chronic_Conditions', 'Allergies'])['Recommended_Medication'].value_counts().unstack()
print(medication_effects)
```

### 14)What insights can be drawn by comparing smoking history, alcohol consumption, and fever severity? 

```python


# Convert Smoking_History and Alcohol_Consumption to binary (if not already done)
df['Smoking_History'] = df['Smoking_History'].map({'Yes': 1, 'No': 0})
df['Alcohol_Consumption'] = df['Alcohol_Consumption'].map({'Yes': 1, 'No': 0})

# Aggregate data
smoking_alcohol_stats = df.groupby('Fever_Severity')[['Smoking_History', 'Alcohol_Consumption']].mean() * 100
print(smoking_alcohol_stats)

# Crosstab for visualization
smoking_crosstab = pd.crosstab(df['Fever_Severity'], df['Smoking_History'])
alcohol_crosstab = pd.crosstab(df['Fever_Severity'], df['Alcohol_Consumption'])

# Visualization: Smoking History
plt.figure(figsize=(8, 6))
sns.barplot(data=smoking_crosstab, ci=None)
plt.title("Smoking History vs Fever Severity")
plt.ylabel("Count")
plt.xlabel("Fever Severity")
plt.show()

# Visualization: Alcohol Consumption
plt.figure(figsize=(8, 6))
sns.barplot(data=alcohol_crosstab, ci=None)
plt.title("Alcohol Consumption vs Fever Severity")
plt.ylabel("Count")
plt.xlabel("Fever Severity")
plt.show()
```

### Pivot Table: Relationship Between Fever Severity and Diet Type

```python
# Creating a pivot table
pivot_table = pd.pivot_table(
    df, 
    values='Heart_Rate',  # Aggregate the mean Heart Rate
    index='Fever_Severity',  # Fever Severity as rows
    columns='Diet_Type',  # Diet Type as columns
    aggfunc='mean',  # Aggregate function: mean
    fill_value=0  # Replace NaN with 0
)

# Print pivot table
print(pivot_table)

# Visualization of the pivot table
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap="viridis")
plt.title("Pivot Table: Mean Heart Rate by Fever Severity and Diet Type")
plt.ylabel("Fever Severity")
plt.xlabel("Diet Type")
plt.show()
```
