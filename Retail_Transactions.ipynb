### 1) Which customer made the highest number of purchases?

```python

import pandas as pd
df=pd.read_csv('C:\\Data Analytics\Workspace\\retail_transactions.csv')
# Count the number of purchases for each customer
customer_purchases = df['Customer_Name'].value_counts()

# Identify the customer with the highest number of purchases
top_customer = customer_purchases.idxmax()
top_purchases = customer_purchases.max()

# Display the result
print(f"The customer with the highest number of purchases is {top_customer}, with {top_purchases} purchases.")
```

### 2)What is the average spending per customer?

```python
# Group by Customer_Name and calculate the total spending per customer
customer_spending = df.groupby('Customer_Name')['Total_Amount'].sum()

# Calculate the average spending per customer
average_spending = customer_spending.mean()

# Display the result
print(f"The average spending per customer is:{average_spending}")
```

### 3)Which customers received the highest discounts?

```python
max_discount = df['Discount_Applied'].max()
customers_with_max_discount = df[df['Discount_Applied'] == max_discount]['Customer_Name'].unique()
print(f"The highest discount received is: ${max_discount:.2f}")
print("Customers who received the highest discount:")
for customer in customers_with_max_discount:
    print(customer)
```

### 4)What is the distribution of payment methods among customers?

```python

payment_method_distribution = df['Payment_Method'].value_counts()
print("Distribution of payment methods among customers:")
print(payment_method_distribution)
```

### 5)Which product generated the highest revenue?

```python
product_revenue = df.groupby('Product_Name')['Total_Amount'].sum()
highest_revenue_product = product_revenue.idxmax()
highest_revenue = product_revenue.max()

# Display the result
print(f"The product that generated the highest revenue is: {highest_revenue_product}")
print(f"Total revenue generated by this product: ${highest_revenue:.2f}")
```

### 6)What is the most frequently purchased product?

```python
product_frequency = df['Product_Name'].value_counts()
most_frequent_product = product_frequency.idxmax()
highest_frequency = product_frequency.max()
print(f"The most frequently purchased product is: {most_frequent_product}")
print(f"Total times it was purchased: {highest_frequency}")
```

### 7)What is the average revenue per transaction?

```python
average_revenue=df['Total_Amount'].mean()
print(f"Average revenue per transaction is:{average_revenue:.2f}")
```

### 8)What percentage of transactions involved discounts?

```python
# Count the total number of transactions
total_transactions = len(df)

# Count the number of transactions with discounts applied
discounted_transactions = len(df[df['Discount_Applied'] > 0])

# Calculate the percentage of transactions with discounts
percentage_with_discounts = (discounted_transactions / total_transactions) * 100

# Display the result
print(f"The percentage of transactions that involved discounts is: {percentage_with_discounts:.2f}%")
```

### 9) Which transaction had the highest total amount?

```python
highest_transaction = df.loc[df['Total_Amount'].idxmax()]
print("Transaction with the highest total amount:")
print(highest_transaction)
```

### 10)Which salesperson contributed the most to sales revenue?

```python
salesperson_revenue = df.groupby('Salesperson_Name')['Total_Amount'].sum()
top_salesperson = salesperson_revenue.idxmax()
highest_revenue = salesperson_revenue.max()
print(f"The salesperson who contributed the most to sales revenue is: {top_salesperson}")
print(f"Total revenue contributed by this salesperson: {highest_revenue:.2f}")
```

### 11)Which store location recorded the highest number of transactions?

```python
store_transaction_count = df['Store_Location'].value_counts()
top_store_location = store_transaction_count.idxmax()
highest_transaction_count = store_transaction_count.max()
print(f"The store location with the highest number of transactions is: {top_store_location}")
print(f"Total transactions recorded at this location: {highest_transaction_count}")
```

### 12)How does the average transaction amount differ across locations?


```python
average_transaction_by_location = df.groupby('Store_Location')['Total_Amount'].mean()
print("Average transaction amount by location:")
print(average_transaction_by_location)
```

### 13)On which date were the most transactions made?

```python
transactions_by_date = df['Date'].value_counts()
most_transactions_date = transactions_by_date.idxmax()
most_transactions_count = transactions_by_date.max()
print(f"The date with the most transactions is: {most_transactions_date}")
print(f"Total transactions on this date: {most_transactions_count}")
```

### 14)What is the average time duration between customer purchases?

```python
df['Date'] = pd.to_datetime(df['Date'])

# Sort the data by Customer_Name and Transaction_Date
data = df.sort_values(by=['Customer_Name', 'Date'])

# Calculate the time difference between consecutive purchases for each customer
data['Time_Difference'] = data.groupby('Customer_Name')['Date'].diff()

# Calculate the average time difference
average_time_difference = data['Time_Difference'].mean()

# Display the result
print(f"The average time duration between customer purchases is: {average_time_difference}")
```

### 15)How much revenue was lost due to discounts?

```python
data['Revenue_Lost'] = data['Discount_Applied']

# Calculate the total revenue lost due to discounts
total_revenue_lost = data['Revenue_Lost'].sum()

# Display the result
print(f"The total revenue lost due to discounts is: ${total_revenue_lost:.2f}")
```

### Are there seasonal trends in sales over the dataset period?

```python
if 'Date' in data.columns and 'Total_Amount' in data.columns:
    df['Date'] = pd.to_datetime(df['Date'])

    # Extract year and month for grouping
    df['YearMonth'] = df['Date'].dt.to_period('M')

    # Aggregate total sales by Year-Month
    sales_trends = df.groupby('YearMonth')['Total_Amount'].sum()

    # Plotting sales trends over time
    plt.figure(figsize=(12, 6))
    sales_trends.plot(kind='line', marker='o', color='green', linestyle='-')
    plt.title('Seasonal Trends in Sales Over Time', fontsize=16)
    plt.xlabel('Time (Year-Month)', fontsize=12)
    plt.ylabel('Total Sales ($)', fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("The dataset does not contain the expected columns: 'Date' and 'Total_Amount'.")
```

### PIE CHART REPRESENTATION

```python
if 'Payment_Method' in df.columns:
    # Count occurrences of each payment method
    payment_distribution = df['Payment_Method'].value_counts()

    # Plotting a pie chart
    plt.figure(figsize=(8, 8))
    payment_distribution.plot(
        kind='pie',
        autopct='%1.1f%%',
        startangle=140,
        colors=plt.cm.Paired.colors,
        wedgeprops={'edgecolor': 'black'}
    )
    plt.title('Distribution of Payment Methods', fontsize=16)
    plt.ylabel('')  # Remove default ylabel
    plt.tight_layout()
    plt.show()
else:
    print("The dataset does not contain the column 'Payment_Method'.")
```

# SUMMARY

### The retail transactions dataset provides a comprehensive overview of how customers interact with a business. It shows which products are the most popular, revealing customer preferences and highlighting items that drive the most revenue. At the same time, some products sell less often, suggesting opportunities for discounts, bundling, or removal from inventory.

### The dataset also uncovers patterns in sales over time, with certain months or seasons showing higher activity—likely due to holidays or promotions—while other periods are quieter. This highlights when businesses should focus their marketing efforts or prepare for higher demand. Discounts are a significant factor in attracting customers but can reduce overall revenue, showing the importance of striking a balance between offering deals and maintaining profitability.

### Insights into payment methods reveal what customers prefer, such as cash, credit cards, or digital wallets, which helps businesses improve convenience and satisfaction. Analysis of store locations shows that some perform much better than others, suggesting opportunities to replicate successful strategies in underperforming areas. Finally, patterns in how often customers shop can guide businesses to implement loyalty programs or send reminders to encourage repeat purchases. Together, these insights help the business plan smarter strategies to boost sales, improve customer experience, and grow revenue.
